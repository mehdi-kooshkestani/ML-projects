```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()   #Standardize features by removing the mean and scaling to unit variance  
                            #Centering and scaling happen independently on each feature
std_scaler = np.array(data_proc[['Income','Kidhome', 'Teenhome', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth','Age', 'Years_customer', 'Total_Expenses', 'Total_Acc_Cmp']]).reshape(-1,19)
        #education and marital status are not in the list above to be standardized
scaler.fit(std_scaler)
data_proc[['Income', 'Kidhome', 'Teenhome', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth','Age', 'Years_customer', 'Total_Expenses', 'Total_Acc_Cmp']] = scaler.transform(std_scaler)
data_proc
```

<span style="color:yellow;"> Finally, down to clustering. In this case Iâ€™ll try using K-means and agglomerative clustering </span>

```python
from sklearn.cluster import KMeans
wcss = []   #Within-Cluster Sum of Square
for i in range(1,5):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)  #k-means++: probability of choosing a point as centroid is directly proportional  
                                                                            # to its distance from the nearest, previously chosen centroid
    kmeans.fit(data_proc)   # data_proc was entered in kmeans process with marital status and education not standardized
    wcss.append(kmeans.inertia_)    #inertia:It is calculated by measuring the distance between each data point and its centroid, squaring this distance,  
                                    # and summing these squares across one cluste
plt.figure(figsize = (10,8))
plt.plot(range(1,5), wcss, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('K-means Clustering')
plt.show()                                    
```

* Ok, 2 is the best clustering based on elbow method

```python
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(data_proc)
label = kmeans.predict(data_proc)
data_segment = data_prep.copy() #data_prep:unstandardized values
data_segment['Segments'] = label
data_segment
```

```python
data_segment.groupby(['Segments']).mean()
```

<span style="color:yellow;">* Segment 0 (few opportunities) = Lower income, have more kids, higher deals purchase, higher web visits, low expense, low accepted campaign</span>
<span style="color:yellow;">* Segment 1 (well-off) = Higher income, less kids, high web, catalog, and store purchases, high expense, higher accepted campaign</span>

```python
data_segment['Labels'] = data_segment['Segments'].map({1:'well-off',0:'fewer-opportunities'})
data_segment
```

<span style="font-size:36px;color:yellow">**Agglomerative**</span>

Agglomerative clustering uses a bottom-up approach, wherein each data point starts in its own cluster. These clusters are then joined greedily, by taking the two most similar clusters together and merging them,ward method is more used to measure the distance between clusters,best point to specify the number of clusters is where most rising in horizontal line occures

```python
from sklearn.cluster import AgglomerativeClustering
agl =  AgglomerativeClustering(n_clusters=2)    #num of  clusters are predefined
agl.fit(data_proc)  #data_proc:standardized values
label = agl.labels_
data_segment_3 = data_prep.copy()   ##data_prep:unstandardized values
data_segment_3['cluster'] = label
data_segment_3.groupby(['cluster']).mean()
```
